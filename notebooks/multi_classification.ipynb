## Set up

```python
doc_path = '../data/download_sf_dataset.parquet'
df_sf_dataset = pd.read_parquet(doc_path)
print(f"shape: {df_sf_dataset.shape}")
print(f"first few rows: {df_sf_dataset.head(5)}")
print(f"columns: {df_sf_dataset.columns}")
```

```python
df_sf_dataset['title'].isnull().sum()
```

```python
df_sf_dataset['summary'].isnull().sum()
```

```python
df_sf_dataset['text'] = df_sf_dataset['title'] + ' ' + df_sf_dataset['summary'] 
df_sf_dataset
```

```python
#remove punctuation, lower case and numbers
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
```

```python
df_sf_dataset['cleaned_text'] = df_sf_dataset['text'].apply(clean_text)
df_sf_dataset
```

```python
def get_sentiment_label(score):
    if score < 0.2:
        sentiment_decision = 'negative'      # negative
    elif score < 0.5:
        sentiment_decision = 'neutral'
    else:
        sentiment_decision = 'positive'
    return sentiment_decision
```
